# Scikit-learn

````python
# Importation des bibliothèques
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt  # Pour afficher le graphique

# Données d'entraînement
ages = np.array([[5], [6], [7], [8], [9]])  # Âges des enfants (2D)
tailles = np.array([105, 115, 120, 125, 130])  # Tailles correspondantes

# Affichage des dimensions (utile pour vérifier les formats)
print("Forme de X (ages) :", ages.shape)
print("Forme de y (tailles) :", tailles.shape)

# Création et entraînement du modèle
modele = LinearRegression()
modele.fit(ages, tailles)

# Affichage de l'équation de la droite
a = modele.coef_[0]       # pente
b = modele.intercept_     # ordonnée à l'origine
print(f"Équation de la droite : y = {a:.2f} * x + {b:.2f}")

# Prédiction pour un enfant de 10 ans
prediction = modele.predict([[10]])
print(f"Taille prédite pour un enfant de 10 ans : {prediction[0]:.2f} cm")

# Tracé du graphique
ages_range = np.array([[4], [10]])  # Intervalle pour la droite
tailles_predites = modele.predict(ages_range)

plt.scatter(ages, tailles, color='blue', label="Données réelles")  # Points observés
plt.plot(ages_range, tailles_predites, color='red', label="Régression linéaire")  # Droite
plt.scatter([10], prediction, color='green', label="Prédiction à 10 ans")  # Point prédit

plt.title("Taille en fonction de l'âge")
plt.xlabel("Âge (années)")
plt.ylabel("Taille (cm)")
plt.legend()
plt.grid(True)
plt.show()
````

## Pandas

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Création manuelle d'un jeu de données
donnees = {
    'Age': [5, 6, 7, 8, 9, 10],
    'Taille_cm': [105, 115, 120, 125, 130, 135],
    'Poids_kg': [18, 20, 22, 25, 27, 30]
}

# 2. Création du DataFrame
df = pd.DataFrame(donnees)

# 3. Affichage des données
print(" Aperçu des données :")
print(df)

# 4. Affichage en barres groupées : taille et poids pour chaque âge
ages = df['Age']
largeur_barre = 0.4
x = np.arange(len(ages))  # Position sur l'axe X

plt.bar(x - largeur_barre/2, df['Taille_cm'], width=largeur_barre, label='Taille (cm)', color='skyblue')
plt.bar(x + largeur_barre/2, df['Poids_kg'], width=largeur_barre, label='Poids (kg)', color='lightgreen')

plt.xticks(x, ages)
plt.xlabel("Âge (années)")
plt.ylabel("Valeur")
plt.title("Taille et poids des enfants par âge")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
import numpy as np  # Importation de la bibliothèque NumPy
import matplotlib.pyplot as plt  # Importation de Matplotlib pour la visualisation

# ------------------------------------------------------
#  1. Tableaux NumPy 1D (vecteurs) et opérations simples
# ------------------------------------------------------

# Création de deux vecteurs (tableaux 1D)
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("=== Opérations élémentaires sur des vecteurs ===")
print("a =", a)
print("b =", b)

# Addition élément par élément : [1+4, 2+5, 3+6]
print("Addition (a + b) :", a + b)  # Résultat : [5 7 9]

# Visualisation graphique des vecteurs
plt.figure(figsize=(6, 4))
plt.plot(a, label='Vecteur a', marker='o')
plt.plot(b, label='Vecteur b', marker='s')
plt.title("Visualisation des vecteurs a et b")
plt.xlabel("Index")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()

print("\n")

# ------------------------------------------------------
#  2. Tableaux NumPy 2D (matrices) et opérations
# ------------------------------------------------------

# Création de deux matrices (tableaux 2D)
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 20], [30, 40]])

print("=== Opérations sur des matrices (tableaux 2D) ===")
print("A =\n", A)
print("B =\n", B)

# Addition de matrices (élément par élément)
print("A + B =\n", A + B)

# Visualisation graphique des matrices
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].imshow(A, cmap='Blues', interpolation='none')
axs[0].set_title("Matrice A")
axs[1].imshow(B, cmap='Oranges', interpolation='none')
axs[1].set_title("Matrice B")

for ax in axs:
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels([0, 1])
    ax.set_yticklabels([0, 1])

plt.tight_layout()
plt.show()

print("\n")

# ------------------------------------------------------
#  3. Fonctions utiles
# ------------------------------------------------------

print("=== Fonctions utiles sur les tableaux ===")
print("Somme des éléments de a :", np.sum(a))         # 6
print("Moyenne des éléments de b :", np.mean(b))      # 5.0
print("Maximum de A :", np.max(A))                    # 4
print("Minimum de B :", np.min(B))                    # 10
print("Forme (dimensions) de A :", A.shape)           # (2, 2)
print("Type des éléments de A :", A.dtype)            # int64 ou int32 selon la machine
````

````python
import numpy as np  # Importation de la bibliothèque NumPy pour les calculs numériques
import matplotlib.pyplot as plt  # Importation de Matplotlib pour la visualisation des données

# ------------------------------------------------------
#  1. Tableaux NumPy 1D (vecteurs) et opérations simples
# ------------------------------------------------------

# Création de deux vecteurs (tableaux 1D)
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Opérations élémentaires sur des vecteurs
print("=== Opérations élémentaires sur des vecteurs ===")
print("a =", a)
print("b =", b)

# Addition élément par élément : [1+4, 2+5, 3+6]
print("Addition (a + b) :", a + b)  # Résultat : [5 7 9]

# Visualisation graphique des vecteurs
plt.figure(figsize=(6, 4))  # Crée une figure avec une taille personnalisée
plt.plot(a, label='Vecteur a', marker='o', color='b')  # Trace le vecteur a avec des cercles
plt.plot(b, label='Vecteur b', marker='s', color='g')  # Trace le vecteur b avec des carrés
plt.title("Visualisation des vecteurs a et b")
plt.xlabel("Index")
plt.ylabel("Valeur")
plt.legend()  # Ajoute une légende pour identifier les courbes
plt.grid(True)  # Ajoute une grille sur le graphique
plt.show()  # Affiche le graphique

print("\n")

# ------------------------------------------------------
#  2. Tableaux NumPy 2D (matrices) et opérations
# ------------------------------------------------------

# Création de deux matrices (tableaux 2D)
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 20], [30, 40]])

# Opérations sur des matrices (élément par élément)
print("=== Opérations sur des matrices (tableaux 2D) ===")
print("A =\n", A)
print("B =\n", B)

# Addition de matrices (élément par élément)
print("A + B =\n", A + B)

# Visualisation graphique des matrices sous forme d'images (avec une échelle de couleurs)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # Crée une figure avec deux sous-graphiques
axs[0].imshow(A, cmap='Blues', interpolation='none')  # Affiche la matrice A avec un cmap bleu
axs[0].set_title("Matrice A")
axs[1].imshow(B, cmap='Oranges', interpolation='none')  # Affiche la matrice B avec un cmap orange
axs[1].set_title("Matrice B")

# Personnalisation des axes
for ax in axs:
    ax.set_xticks([0, 1])  # Définit les positions des ticks sur l'axe X
    ax.set_yticks([0, 1])  # Définit les positions des ticks sur l'axe Y
    ax.set_xticklabels([0, 1])  # Étiquette les ticks de l'axe X
    ax.set_yticklabels([0, 1])  # Étiquette les ticks de l'axe Y

plt.tight_layout()  # Ajuste l'espacement des sous-graphiques
plt.show()  # Affiche le graphique

print("\n")

# ------------------------------------------------------
#  3. Fonctions utiles sur les tableaux NumPy
# ------------------------------------------------------

print("=== Fonctions utiles sur les tableaux ===")
print("Somme des éléments de a :", np.sum(a))         # 6
print("Moyenne des éléments de b :", np.mean(b))      # 5.0
print("Maximum de A :", np.max(A))                    # 4
print("Minimum de B :", np.min(B))                    # 10
print("Forme (dimensions) de A :", A.shape)           # (2, 2)
print("Type des éléments de A :", A.dtype)            # int64 ou int32 selon la machine

print("\n")

# ------------------------------------------------------
#  4. Visualisation graphique : Évolution dans le temps
# ------------------------------------------------------

# Données pour l'évolution dans le temps
temps = [0, 1, 2, 3, 4]  # Liste représentant le temps
valeurs = [0, 2, 4, 6, 8]  # Liste représentant les valeurs associées à chaque temps

# Tracer l'évolution des valeurs dans le temps
plt.figure(figsize=(6, 4))  # Crée une figure avec une taille personnalisée
plt.plot(temps, valeurs, marker='o', color='b')  # Trace la courbe avec un cercle pour chaque point
plt.xlabel('Temps')  # Ajoute un label à l'axe X
plt.ylabel('Valeur')  # Ajoute un label à l'axe Y
plt.title('Évolution dans le temps')  # Ajoute un titre au graphique
plt.grid(True)  # Ajoute une grille pour mieux lire le graphique
plt.show()  # Affiche le graphique
````

**TensorFlow**

````python
import numpy as np  # Importation de NumPy pour générer des données aléatoires
import tensorflow as tf  # Importation de TensorFlow pour construire et entraîner le modèle
from tensorflow.keras.models import Sequential  # Importation du modèle séquentiel de Keras
from tensorflow.keras.layers import Dense  # Importation des couches Dense pour les couches entièrement connectées

# ------------------------------------------------------
#  1. Création du modèle avec Keras
# ------------------------------------------------------

# Initialisation du modèle séquentiel (couches empilées)
model = Sequential()

# Ajout de la première couche dense :
# - 10 neurones, fonction d'activation ReLU et spécification de l'entrée avec 8 caractéristiques
model.add(Dense(10, activation='relu', input_shape=(8,)))

# Ajout de la deuxième couche dense (1 neurone pour la sortie binaire)
model.add(Dense(1, activation='sigmoid'))

# Compilation du modèle :
# - Utilisation de l'optimiseur Adam et de la fonction de perte 'binary_crossentropy' pour une classification binaire
model.compile(optimizer='adam', loss='binary_crossentropy')

# ------------------------------------------------------
#  2. Génération de données fictives pour l'entraînement
# ------------------------------------------------------

# Générer des données d'exemple (10 échantillons, chaque échantillon ayant 8 caractéristiques)
X_train = np.random.rand(10, 8)  # 10 exemples, chacun avec 8 caractéristiques

# Générer des labels binaires (0 ou 1) associés aux échantillons
y_train = np.random.randint(0, 2, size=(10, 1))  # 10 labels binaires (0 ou 1)

# ------------------------------------------------------
#  3. Entraînement du modèle
# ------------------------------------------------------

# Entraînement du modèle avec les données générées
model.fit(X_train, y_train, epochs=10, batch_size=2)

# ------------------------------------------------------
#  4. Évaluation et prédictions (facultatif)
# ------------------------------------------------------

# Prédiction sur de nouvelles données
X_new = np.random.rand(3, 8)  # Générer 3 nouveaux exemples pour tester
predictions = model.predict(X_new)  # Prédire avec le modèle entraîné

print("Prédictions sur de nouvelles données :", predictions)
````
